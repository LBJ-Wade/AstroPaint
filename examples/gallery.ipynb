## Spongy Terror

```python
catalog_sponge = Catalog()
catalog_sponge.generate_random_shell(n_tot=20000)
```

```python
catalog_sponge.cut_lon_lat([0,30], [0,30])
```

```python
canvas_sponge = Canvas(catalog_sponge, nside=1024, R_times=4)
```

```python
def sponge(R, R_200c, M_200c, x, y):
    
    layer = np.log(M_200c)*np.exp(-(2*(R-1)/R_200c)**2)*x
    layer -= np.log(M_200c)**2*np.exp(-(4*R/R_200c)**2)
    return layer
```

```python
painter = Painter(sponge)
```

```python
R = np.linspace(0,5,100)
painter.plot_template(R,catalog_sponge, [0,10,100,200])
```

```python
canvas_sponge.clean()
painter.spray(canvas_sponge)
```

```python
plt.figure(dpi=150)
canvas_sponge.cmap=cm.Greys_r
canvas_sponge.show_map("cart", lonra=[-3,33], latra=[-3,33],fig=1, title="", cbar=False )
```

## Burning Twilight

```python
catalog_twilight = Catalog()
catalog_twilight.generate_random_shell(n_tot=10000, shell_radius=350)
```

```python
nside=1024
canvas_twilight = Canvas(catalog_twilight, nside=nside, R_times=10)
```

```python
def twilight(R,  R_200c, x, y, z):

    layer  = np.exp(-(R/R_200c)/4)*(x+y+z)
    layer  += np.exp(-(R/R_200c*1.5)**2)*(x-y-z) 
    
    return layer
```

```python
painter = Painter(twilight)
```

```python
R = np.linspace(0,20,100)
painter.plot_template(R, catalog_twilight, [0,100,1000])
```

```python
canvas_twilight.clean()
painter.spray(canvas_twilight)
```

```python
plt.figure(dpi=150)
canvas_twilight.show_map("cart", lonra=[-40,40],latra=[-40,40], min=-100, max=1000, fig=1, title="", cbar=False)
#plt.savefig("twilight.png", dpi=150, bbox_inches="")
```
